```python
import os
```

```python
dataset_dir = 'dataset'
print(dataset_dir)
```

```python

def verifier(dataset_dir):
    # Parcourt tous les dossiers du répertoire dataset
    for folder_name in os.listdir(dataset_dir):
        folder_path = os.path.join(dataset_dir, folder_name)
        
        # Vérifie si c'est un dossier
        if os.path.isdir(folder_path):
            # Récupère toutes les images avec les extensions .jpg, .jpeg, .png, .webp
            images = [f for f in os.listdir(folder_path) if f.endswith(('.jpg', '.jpeg', '.png', '.webp'))]
            
            # Renomme les fichiers dans chaque dossier
            for index, image_name in enumerate(images, start=1):
                new_name = f"{folder_name}{index}{os.path.splitext(image_name)[1]}"
                new_path = os.path.join(folder_path, new_name)
                old_path = os.path.join(folder_path, image_name)

                # Vérifie si le fichier de destination existe déjà
                if os.path.exists(new_path):
                    # Si le fichier existe déjà, on génère un autre nom unique
                    base_name, ext = os.path.splitext(new_name)
                    counter = 1
                    while os.path.exists(new_path):
                        new_name = f"{base_name}_{counter}{ext}"
                        new_path = os.path.join(folder_path, new_name)
                        counter += 1

                # Renommage du fichier
                os.rename(old_path, new_path)
                print(f"Renommé {old_path} en {new_path}")
```

```python
verifier(dataset_dir)
```
